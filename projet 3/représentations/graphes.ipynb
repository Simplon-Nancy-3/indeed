```python
# Répartion des salaires suivant les différents contrats
contract = ['contract_temps plein', 'contract_cdd','contract_indépendant', 'contract_stage', 'contract_temps partiel',
            'contract_intérim', 'contract_commission', 'contract_contrat pro','contract_cdi']

def g(x):
    print("\n", df[[x,'salary_min', 'salary_max', 'salary_mean']].groupby([x],as_index=False).mean())
    sns.swarmplot(x=x,y= 'salary_mean', data=df)
interact(g, x=contract)
```

```python
# comparaison du salaire entre les différents métiers
from itertools import cycle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')

def g(x):
    def make_kde(*args,**kwargs):
        sns.kdeplot(*args, cmap=next(make_kde.cmap_cycle), **kwargs)

    make_kde.cmap_cycle = cycle(('coolwarm', 'Wistia', 'icefire'))
    our_plot = sns.PairGrid(df, vars=('salary_min', 'salary_max', 'salary_mean'), hue=x, dropna=True)

    our_plot.map_diag(sns.kdeplot)
    our_plot.map_lower(make_kde)
    our_plot.map_upper(plt.scatter)
    

    our_plot.fig.suptitle("Répartition des salaires entre les différents métiers", fontsize=12)
    our_plot.add_legend()

metiers = ['query_business+intelligence', 'query_data+scientist','query_data+analyst']
''' Entrée listée '''
interact(g, x=metiers)
```
