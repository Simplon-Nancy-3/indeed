# Faire un tri des colonnes

### Méthode de Lasso

```python
# Methode de Lasso: 
def column_selection_Lasso(df, col, sep=0):
    import numpy as np
    from sklearn.linear_model import LassoCV
    X = df.drop(col, axis=1)
    y = df[col]
    reg = LassoCV()
    reg.fit(X, y)
    coef = pd.Series(reg.coef_, index = X.columns)
    imp_coef = coef.sort_values()
    features = pd.DataFrame(imp_coef.where(np.abs(imp_coef) > sep).dropna()).transpose().columns
    y_ = pd.Index([col])
    features = features.append(y_)
    return df[features]

df_c = df_salaire[['query_developpeur','query_data+analyst',
                'query_business+intelligence','query_data+scientist',
                'contract_contrat pro','contract_intérim','contract_temps plein',
                'contract_cdi','contract_commission','contract_temps partiel',
                'contract_apprentissage','contract_cdd','contract_stage','contract_indépendant',
              'salary_mean']]
column_selection_Lasso(df_c, 'salary_mean')
```

### Méthode de tri avec les correlations

```python
# Tri suivant les correlations avec la colonne concernée
def abs_corr_sup(df, col, k):    
    ''' Colonnes ayant une correlation tq |corr avec col| > k '''
    cor_target = abs(df.corr()[col])
    return df[dict(cor_target[cor_target>k]).keys()]
       
abs_corr_sup(df_c, 'salary_mean',0.05)

#sns.heatmap(df.corr(), annot=True, cmap=plt.cm.Reds)
#plt.show()
```
